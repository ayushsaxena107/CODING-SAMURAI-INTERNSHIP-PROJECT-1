```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/Ayush/Desktop/CODING SAMURAI INTERNSHIP/Project 1 (Exploratory Data Analysis (EDA) on Titanic)/train.csv")
```

```python
#Visualize the Data
```

```python
#1)Survival Count

sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
#2) Survival by Gender

sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
# 3) Survival by Class

sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Class')
plt.show()
```

```python
# 4) Age Distribution

sns.histplot(df['Age'].dropna(), bins=30)
plt.title('Age Distribution')
plt.show()
```

```python
# 5) Correlation Heatmap

numeric_df = df.select_dtypes(include=[np.number])  # Keep only numeric columns
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Analyze Relationships


df.groupby('Sex')['Survived'].mean()
df.groupby('Pclass')['Survived'].mean()
df.groupby('Embarked')['Survived'].mean()
```

```python
## Insights

#- The overall survival rate was approximately 38%.
#- Females had a survival rate of 74%, compared to just 19% for males.
#- First-class passengers had the highest survival rate (63%), while third-class had the lowest (24%).
#- Younger children (under 10) had a higher survival rate, showing that “women and children first” was followed to some extent.
```
